##### Importing Modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt    # library to enhance the interactive visualization 
import matplotlib.colors           # module for converting numbers or color arguments to RGB or RGBA
import matplotlib.collections      # classes for efficient drawing of large collection of objects
import seaborn as sns              # library used to draw statistical drawings and graphics
%matplotlib inline
```

##### Defining Functions

```python
def line_plot(data, xaxis, yaxis, xlbl, ylbl, lbl, color):
    
    """This is the docstring of the function line_plot.
    It takes the dataframe along with what to plot on  X axis, Y axis. 
    The function call passes the label & title information 
    based on the user data or what data needs to be plot.  """
    
    plt.figure(figsize=(8, 4), tight_layout=True)
    plt.figure()
    plt.plot(data[xaxis], data[yaxis], color, label=lbl)
    plt.legend()
    plt.xlabel(xlbl)
    plt.ylabel(ylbl)
    
    plt.show()
    

def Bar_plot_inverted(data):
    
    """This is the docstring of the function Bar_plot_inverted.
    
    It describes the dataframe, X axis, Y axis to plot.
    The function call passes the labels & title details 
    based on the user data to be plot and size of the figure defined  """
    
    plt.figure(figsize=(12, 6), tight_layout=True)
    colors = sns.color_palette('deep')
    plt.barh(data['Country'], data['Corruption'], color=colors[:5])
    plt.xlabel('Corruption')
    plt.ylabel('Country')
    plt.title('Corruption Trend')
    
    plt.show()
    

def Pie_chart_sub(data1, data2, lbl1, lbl2):
    
    """This is the docstring of the function Pie chart.
    
    This function takes the dataframe 1 & dataframe 2 along with corresponding labels
    as the inputs. Data frame 1 is plotted on the axis 0 & dataframe 2 
    plotted on the axis 1. Along with the data frames, we formatted the pie chart
    properties like colors,shadow,thickness."""
    
    #Creating the sub plots
    fig, ax = plt.subplots(1, 2, figsize=(12, 5), tight_layout=True)
    
   #Creating the first axis plot
    colors = sns.color_palette('pastel')      
    

    explode_list = [0, 0, 0, 0, 0.2]
    
    #setting axis for piechart
    ax[0].pie(data1['Happiness_value'], labels=data1['Country'], 
              autopct='%.0f %%', explode=explode_list, pctdistance=.7, 
              colors=colors, shadow=True)

    ax[0].set_title(lbl1, weight='bold')

    #defining the second Axis Plot
    colors = sns.color_palette('deep')


    explode_list = [0, 0, 0, 0, 0.2]

    ax[1].pie(data2['Happiness_value'], labels=data2['Country'],
    autopct='%.0f %%', explode=explode_list, pctdistance=.7,
    colors=colors, shadow=True)

    ax[1].set_title(lbl2, weight='bold')

    plt.show()
    
def trend_plot(df_trend, xlbl, ylbl, plt_title, cont1, cont2, cont3, cont4):
     
    """This is the docstring of the function trend_plot.
    
    This is a custom made plot function. It is the merged form of multiple line plots.
    This is being used for analyzing the trend over the years with multiple variables 
    with list of  countries on the additional legend box. Along with the data frames,
    we are formatting the graph properties like colors ,thickness. """
    
    #plotting the figure
    plt.figure(figsize=(10, 6), tight_layout=True)
    
    #customizing the plot
    plt.plot(df_trend, 'o-', linewidth=4)
    
    plt.xticks([2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020])
    plt.xlabel(xlbl)
    plt.ylabel(ylbl)
    plt.title(plt_title)
    plt.legend(title='Countries', title_fontsize=13, labels=[cont1,cont2,cont3,cont4])
    
    plt.show()
    

def violin_plt_mult(data1, data2, data3, data4, xlbl1,
                    xlbl2, xlbl3, xlbl4, titl1, titl2,
                    titl3, titl4):
     
    """This is the docstring of the function violin plot.
    
    This function takes four different data frames 1,2,3 and 4 with corresponding labels & titles 
    as the inputs.  This violin plot written in the sense of plotting multiple data variables
    on different vertical axis. Dataframe 1 is plotted on the axis 0 and dataframe 2
    plotted on the axis 1 and so on. Along with the data frames, setting the graph 
    properties like colors, thickness, size of the plot etc.  """
    
    colors_list = ['#78C850', '#F08030',  '#6890F0',  '#A8B820',  '#F8D030', 
                   '#E0C068', '#C03028', '#F85888', '#98D8D8']

    fig = plt.figure(figsize=(6, 6))
    gs = fig.add_gridspec(1, 4)
    ax = fig.add_subplot(gs[0, 0])

    sns.violinplot(data=data1, palette=colors_list)
    ax.set_xlabel(xlbl1)
    ax.set_title(titl1)

    ax = fig.add_subplot(gs[0, 1])
    sns.violinplot(data=data2)
    ax.set_xlabel(xlbl2)
    ax.set_title(titl2)

    ax = fig.add_subplot(gs[0, 2])
    sns.violinplot(data=data3, palette=colors_list)
    ax.set_xlabel(xlbl3)
    ax.set_title(titl3)
    
    ax = fig.add_subplot(gs[0, 3])
    sns.violinplot(data=data4, palette=colors_list)
    ax.set_xlabel(xlbl4)
    ax.set_title(titl4)
    
    fig.tight_layout()
    plt.show()

    
def Scatter_plot(df, xaxis, yaxis ,lblx, lbly):
     
    """This is the docstring of the function Scatter plot.
    
    It takes the dataframe along with the X axis, Y axis to plot. 
    The function call passes the label & title information 
    based on the user data or what data needs to be plot and 
    setting the color properties of the plot. """
    
    plt.figure(figsize=(10, 6), tight_layout=True)
    
    ax = sns.scatterplot(data=df, x=xaxis, y=yaxis,
                         hue='Country', palette='dark', s=400)
    
    # setting axis GDP vs Happiness
    ax.set(xlabel=lblx, ylabel=lbly)  
    ax.set_title('Corruption vs Freedom Analysis')
    
    plt.show()
      
```

##### Reading & Pre-processing the Data Set received from the World Happiness Survey

```python
#Reading the World Happiness Survery Dataset for the year 2021
df_init = pd.read_csv("world-happiness-report-2021.csv") 


#Reading the World Happiness Survery Dataset from 2008-2020
df_init_full = pd.read_csv("world_happiness_report.csv") 
```
